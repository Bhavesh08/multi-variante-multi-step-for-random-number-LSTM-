```python
import pandas as pd
```

```python
x = pd.DataFrame()
```

```python
x['obj'] = [i for i in range(10)]
x['obj1'] = [e for e in range(50,60)]
```

```python
x = x.values
```

```python
df = pd.DataFrame(x)
```

```python
n_vars = 1 if type(df) is list else df.shape[1]
n_in = 2
n_out = 2
col,names = list(),list()
for p in range(n_in ,0, -1):
    r = df.shift(p)
    col.append(r)
    names += ['var%d(t-%d)'%(j+1,p)for j in range(n_vars)]
for q in range(0,n_out):
    g = df.shift(-q)
    col.append(g)
    if q == 0:
        names += ['var%d(t)'%(l+1)for l in range(n_vars)]
    else:
        names += ['var%d(t+%d)'%(y+1,q)for y in range(n_vars)]
```

```python
print(col,names)
```

```python
agg = pd.concat(col,axis = 1)
```

```python
agg.columns = names
```

```python
agg.dropna(inplace = True)
```

```python
agg.shape
```

```python
agg
```

```python
X = agg.drop(['var1(t)','var2(t)','var1(t+1)','var2(t+1)'],axis = 1)
y = agg[['var1(t)','var2(t)','var1(t+1)','var2(t+1)']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 8)
```
